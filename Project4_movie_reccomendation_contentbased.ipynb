```python
import ast
```

```python
def paste(a):
    list = []
    for i in ast.literal_eval(a):
        list.append(i['name'])

    return list
```

```python
def convert(b):
    l = []
    counter = 0
    for i in ast.literal_eval(b):
        if counter !=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
        
```

```python
def convert2(c):
    l2 = []
    for i in ast.literal_eval(c):
        if i['job'] == 'Director':
            l2.append(i['name'])
            break

    return l2
```

NLTK

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
['loved','loving','lovely','love']
```

```python
ps.stem('loving')
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return ' '.join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

VECTORIZATION

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
#SPARSE MATRIX
# A sparse matrix is a matrix that contains a large number of zero elements. In contrast, a dense matrix has a significant number of non-zero elements.
cv.fit_transform(df['tags'])
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
np.set_printoptions(threshold=np.inf)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#OPTIONAL
from scipy.spatial.distance import pdist, squareform
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
        
```

```python
recommend('Batman Begins')
```
